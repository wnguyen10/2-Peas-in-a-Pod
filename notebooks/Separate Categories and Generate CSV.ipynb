```python
import pandas as pd
```

```python
df = pd.read_csv("trunc_metadata.csv")
```

```python
df
```

```python
df["category"] = ""
df["subcategory"] = ""
```

```python
df
```

```python
for i, category in enumerate(df.iat[2, 6].strip('][').split(', ')):
    category = category.strip("'")
    if i == 0:
        df.iat[2, 8] = category
    else:
        df.iat[2, 9] = category
```

```python
for i in range(len(df)):
    category_string = df.iat[i, 6]
    category_list = category_string.strip('][').split(', ')
    for j, category in enumerate(category_list):
        category = category.strip("'")
        if j == 0:
            df.iat[i, 8] = category
        else:
            df.iat[i, 9] = category
```

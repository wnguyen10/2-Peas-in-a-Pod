```python
df = pd.read_csv("/content/drive/MyDrive/4300 Project/shows.csv")
```

```python
categories = df["category"]

cat_map = {}
for c in categories:
  if c not in cat_map:
    cat_map[str(c)] = 0
  cat_map[str(c)] += 1
# console.log(len(cat_map))

# categories = sorted(cat_map.items(), key =  lambda x: x[1], reverse = True)[:10]

cat_map = dict(filter(lambda x: x[1] > 100, cat_map.items()))

print(len(cat_map))
x = list(cat_map.keys())
y = list(cat_map.values())

# plt.figure(figsize=(20, 10))

fig, ax = plt.subplots(figsize=(20,10))
for tick in ax.get_xticklabels():
    tick.set_rotation(90)

plt.bar(x, y)



```

```python
exp = df["explicit"]

exp_count = 0
non_count = 0

for x in exp:
  if x:
    exp_count += 1
  else:
    non_count += 1

plt.pie([exp_count, non_count], labels=["Explicit", "Non-explicit"], colors = ["black", "yellow"], shadow = True)
```

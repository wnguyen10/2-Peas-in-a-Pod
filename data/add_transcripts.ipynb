```python
import pandas as pd
import numpy as np
import os
import sys
import json
```

```python
# Read initial metadata TSV
df = pd.read_table("metadata.tsv")
```

```python
df
```

```python
# View columns
for col in df.columns:
    print(col)
```

```python
# Folder containing all transcript JSONs
transcript_folder_path = os.path.join(os.getcwd() , 'spotify-podcasts-2020/podcasts-transcripts')
```

```python
# Add empty transcript column
df['transcript'] = None
```

```python
for root, dirs, files in os.walk(transcript_folder_path, topdown = False):
    
    for name in files:
        if name.endswith(".json"):
            filepath = os.path.join(root, name)
            f = open(filepath)
            data = json.load(f)
            
            # SHOW_FILENAME_PREFIX
            show_prefix = os.path.split(os.path.split(filepath)[0])[1]
            
            # EPISODE_FILENAME_PREFIX
            episode_prefix = os.path.splitext(name)[0]
            
            full_transcript = ""
            
            # Iterate through all JSONs in subdirectory and concatenate transcripts
            for d in data['results']:
                if "alternatives" in d:
                    if "transcript" in d["alternatives"][0]:
                        transcript = d['alternatives'][0]['transcript']
                        full_transcript += transcript
             
            # Get row index based on episode filename prefix (unique per row)
            row_index = df[df['episode_filename_prefix'] == episode_prefix].index.item()
            
            # Transcript column is at index 12, replace transcript cell with full transcript
            df.iat[row_index, 12] = full_transcript
```

```python
df.to_csv("metadata_with_transcripts.csv", index=False)
```

```python
updated_df = english_df.drop(["language", "episode_uri", "episode_name", "show_filename_prefix", "episode_filename_prefix"], axis = 1)
```

```python
updated_df
```

```python
# Add space after each transcript to allow for string concatenation
updated_df['transcript'] = updated_df['transcript'] + " "
```

```python
agg_functions = {"show_uri": "first", "show_description": "first", "publisher": "first", "rss_link": "first", "episode_description": "sum", "duration": "mean", "transcript": "sum"}
```

```python
updated_df = updated_df.groupby(updated_df['show_name'], as_index = False).aggregate(agg_functions)
```

```python
# Remove podcast with blank name
updated_df = updated_df[updated_df["show_name"] != " "]
```

```python
updated_df
```

```python
updated_df.iloc[0]['transcript']
```

```python
updated_df = updated_df.rename(columns = {"episode_description": "ep_descriptions", "duration": "avg_duration", "transcript": "ep_transcripts"})
```

```python
updated_df.nunique()
```

```python
df_final = updated_df.copy()
```

```python
# Remove emojis and other non-ascii characters from dataframe
df_final = df_final.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
```

```python
# With emojis removed, remove all shows with empty show_name, show_description, publisher, and/or ep_descriptions
df_final = df_final[df_final["show_name"] != ""]
df_final = df_final[df_final["show_description"] != " "]
df_final = df_final[df_final["publisher"] != ""]
df_final = df_final[df_final["ep_descriptions"] != " "]
```

```python
df_final.to_csv("preprocessed_shows.csv")
```
